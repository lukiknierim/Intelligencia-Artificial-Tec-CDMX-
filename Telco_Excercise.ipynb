```python
import pandas as pd
import numpy as np




df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
pd.set_option('display.max_columns', None)

df.head()
```

```python
df["TotalCharges"] = (
    pd.to_numeric(df["TotalCharges"], errors="coerce")
      .fillna(0)
)
```

```python
print("Tenure")
print("--------------------------")
print("Mode of tenure is: ", df["tenure"].mode())
print("Mean of tenure is: ", df["tenure"].mean())
print("Median of tenure is: ", df["tenure"].median())


print(




)

print("MonthlyCharges")
print("--------------------------")
print("Mode of MonthlyCharges is: ", df["MonthlyCharges"].mode())
print("Mean of MonthlyCharges is: ", df["MonthlyCharges"].mean())
print("Median of MonthlyCharges is: ", df["MonthlyCharges"].median())



print("TotalCharges")
print("--------------------------")
print("Mode of TotalCharges is: ", df["TotalCharges"].mode())
print("Mean of TotalCharges is: ", df["TotalCharges"].mean())
print("Median of TotalCharges is: ", df["TotalCharges"].median())
```

```python
import matplotlib.pyplot as plt
Tenure = pd.to_numeric(df["tenure"], errors="coerce").dropna()
MonthlyCharges = pd.to_numeric(df["MonthlyCharges"], errors="coerce").dropna()

plt.figure(); plt.hist(Tenure, bins=30); plt.title("Tenure"); plt.xlabel("Value"); plt.ylabel("Count"); plt.show()
plt.figure(); plt.hist(MonthlyCharges, bins=30); plt.title("MonthlyCharges"); plt.xlabel("Value"); plt.ylabel("Count"); plt.show()
```

```python

churn_norm = (
    df["Churn"]
    .astype("string")
    .str.strip()
    .str.lower()
)

df_churn    = df[churn_norm.eq("yes")].copy()
df_nochurn  = df[churn_norm.eq("no")].copy()

df_churn.head()
```

```python

print("Churn")
print("--------------------------")
print("Median of MonthlyCharges is: ", df_churn["MonthlyCharges"].median())
print("Median of Tenure is: ", df_churn["tenure"].median())



print(

)
print("No Churn")
print("--------------------------")
print("Median of MonthlyCharges is: ", df_nochurn["MonthlyCharges"].median())
print("Median of Tenure is: ", df_nochurn["tenure"].mean())
```

```python
import matplotlib.pyplot as plt
Churn = pd.to_numeric(df_churn["MonthlyCharges"], errors="coerce").dropna()
Nochurn = pd.to_numeric(df_nochurn["MonthlyCharges"], errors="coerce").dropna()

plt.figure(); plt.hist(Churn, bins=30); plt.title("Churn"); plt.xlabel("Value"); plt.ylabel("Count"); plt.show()
plt.figure(); plt.hist(Nochurn, bins=30); plt.title("NoChurn"); plt.xlabel("Value"); plt.ylabel("Count"); plt.show()
```
