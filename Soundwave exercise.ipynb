```python
import librosa
from IPython.display import Audio
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
```

```python
data_dog, sampling_rate_dog = librosa.load(r"/content/barking-large-and-small-dog-290711.wav")
```

```python
data_dog
```

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(data_dog, sr=sampling_rate_dog)
plt.show()
```

```python
Audio(data=data_dog, rate=sampling_rate_dog)
```

```python
Audio(data=data_dog, rate=sampling_rate_dog*2)
```
